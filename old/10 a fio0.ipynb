```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['wine.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```python
#descartar as colunas não necessárias do dataframe
df = df.drop(columns={'Magnesium','Total_phenols','Proanthocyanins','Hue','OD280_OD315','Proline'})
```

```python
import plotly.express as px

fig = px.parallel_coordinates(df, color="Color_intensity" , labels={"Color_intensity":"Intensidade da Cor" ,"Alcohol": "Álcool",
                  "Malic_acid": "Ácido málico", "Ash": "Cinzas", "Alcalinity_ash" : "Alcalinidade das cinzas", "Total_phenols" : "Fenóis Totais",
                  "Flavanoids": "Flavanoides", "Nonflavanoid_phenols": "Fenóis não flavorizados" },
                    #color_continuous_scale=px.colors.diverging.Tealrose, color_continuous_midpoint=2
                    )
fig.show()
```

```python
fig = px.scatter(df, x = "Malic_acid", y = "Alcohol", size="Flavanoids", color = "Color_intensity")
fig.show()
```
